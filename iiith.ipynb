```python
# Step 3: Train the model
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # Or use yolov8s.pt if you want slightly better accuracy

model.train(
    data='/content/drive/MyDrive/Cholec80_YOLO/data.yaml',  # path to your data.yaml
    epochs=50,
    imgsz=640,
    batch=16
)
```

```python
# Replace with the path to one of your original Cholec80 videos
video_path = "/content/drive/MyDrive/Cholec80_Project/videos/video01.mp4"

# Run detection
model.predict(
    source=video_path,
    conf=0.25,              # Confidence threshold
    save=True,              # Save output video
    save_txt=False,         # Save detection results as txt
    save_conf=True,         # Save confidence score
    imgsz=640,              # Image size
    show=False              # Don't display in Colab
)
```
